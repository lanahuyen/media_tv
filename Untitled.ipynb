```python
import altair as alt
import pandas as pd
```

```python
# Load cleaned dataset for streaming platforms
streaming_df = pd.read_csv('data/cleaned_MoviesOnStreamingPlatforms.csv')

# Prepare data for visualization: Count titles for each platform
platform_counts = streaming_df[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum().reset_index()
platform_counts.columns = ['Platform', 'Title Count']

# Create a bar chart
bar_chart = alt.Chart(platform_counts).mark_bar().encode(
    x=alt.X('Platform', sort=None, title='Streaming Platform'),
    y=alt.Y('Title Count', title='Number of Titles'),
    color='Platform'
).properties(
    title='Number of Titles Available on Streaming Platforms',
    width=600,
    height=400
)

bar_chart
```

```python
# Load cleaned subscription price history data
price_history_df = pd.read_csv('data/cleaned_AllServicesPriceHistory.csv')

# Create a line chart for subscription costs over time
line_chart = alt.Chart(price_history_df).mark_line().encode(
    x=alt.X('Date:T', title='Date'),
    y=alt.Y('Subscription Price', title='Price (USD)'),
    color='Streaming Service'
).properties(
    title='Subscription Cost Over Time for Major Streaming Services',
    width=800,
    height=400
)

line_chart
```

```python
# Load cleaned Netflix Userbase data
userbase_df = pd.read_csv('data/cleaned_NetflixUserbase.csv')

# Prepare data for pie chart: Count users by subscription type
subscription_counts = userbase_df['Subscription Type'].value_counts().reset_index()
subscription_counts.columns = ['Subscription Type', 'Count']

# Create a pie chart
pie_chart = alt.Chart(subscription_counts).mark_arc().encode(
    theta='Count:Q',
    color='Subscription Type:N',
    tooltip=['Subscription Type:N', 'Count:Q']
).properties(
    title='User Distribution by Subscription Type'
)

pie_chart
```

```python
# Prepare data: Count users by device
device_counts = userbase_df['Device'].value_counts().reset_index()
device_counts.columns = ['Device', 'Count']

# Create a bar chart for device usage
device_bar_chart = alt.Chart(device_counts).mark_bar().encode(
    x=alt.X('Device:N', sort='-y', title='Device'),
    y=alt.Y('Count:Q', title='Number of Users'),
    color='Device:N',
    tooltip=['Device', 'Count']
).properties(
    title='Distribution of Devices Used by Netflix Users',
    width=600,
    height=400
)

device_bar_chart
```

```python
# Group data by country and calculate total monthly revenue
revenue_by_country = userbase_df.groupby('Country')['Monthly Revenue'].sum().reset_index()

# Create a bar chart for monthly revenue by country
revenue_bar_chart = alt.Chart(revenue_by_country).mark_bar().encode(
    x=alt.X('Country:N', sort='-y', title='Country'),
    y=alt.Y('Monthly Revenue:Q', title='Total Monthly Revenue (USD)'),
    color='Country:N',
    tooltip=['Country', 'Monthly Revenue']
).properties(
    title='Total Monthly Revenue by Country',
    width=800,
    height=400
).configure_axis(
    labelAngle=-45
)

revenue_bar_chart
```

```python
# Create a box plot for age distribution by subscription type
age_box_plot = alt.Chart(userbase_df).mark_boxplot().encode(
    x=alt.X('Subscription Type:N', title='Subscription Type'),
    y=alt.Y('Age:Q', title='Age of Users'),
    color='Subscription Type:N',
    tooltip=['Subscription Type', 'Age']
).properties(
    title='Age Distribution by Subscription Type',
    width=600,
    height=400
)

age_box_plot
```

```python
# Create a scatter plot for monthly revenue vs age
revenue_age_scatter = alt.Chart(userbase_df).mark_circle(size=60).encode(
    x=alt.X('Age:Q', title='Age'),
    y=alt.Y('Monthly Revenue:Q', title='Monthly Revenue (USD)'),
    color='Subscription Type:N',
    tooltip=['Age', 'Monthly Revenue', 'Subscription Type']
).properties(
    title='Monthly Revenue vs Age of Netflix Users',
    width=600,
    height=400
)

revenue_age_scatter
```
