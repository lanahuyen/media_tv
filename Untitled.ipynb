```python
import pandas as pd
import altair as alt

# Mock data for the evolution of media consumption
data = {
    'Year': [1920, 1950, 1980, 2000, 2010, 2020],
    'Milestone': [
        'Radio becomes popular',
        'Television introduction',
        'Cable TV popularity rises',
        'Digital media & DVDs',
        'Birth of streaming services',
        'Streaming dominates media consumption'
    ],
    'Type': [
        'Radio', 'Television', 'Cable TV', 'Digital Media', 'Streaming', 'Streaming'
    ]
}

df = pd.DataFrame(data)

# Create a timeline chart
timeline_chart = alt.Chart(df).mark_circle(size=200).encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.value(0),
    color=alt.Color('Type:N', scale=alt.Scale(scheme='category20b'), legend=alt.Legend(title='Media Type')),
    tooltip=['Year', 'Milestone'],
).properties(
    title='Evolution of Media Consumption',
    width=800,
    height=200
)

# Add text annotations for milestones
text = timeline_chart.mark_text(
    align='left',
    baseline='middle',
    dx=7  # Adjust horizontal position
).encode(
    text='Milestone'
)

# Combine the timeline with text annotations
final_timeline_chart = timeline_chart + text

final_timeline_chart
```
