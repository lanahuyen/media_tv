```python
from pymongo import MongoClient
import pandas as pd

# Connect to MongoDB (default connection)
client = MongoClient('mongodb://localhost:27017/')

# Create or switch to the database
db = client['media_behavior']

# Create or switch to collections
cable_tv_collection = db['cable_tv_data']  # Ensure this variable is defined

# Load and clean cable TV data
cable_tv_df = pd.read_json('data/cabel_tv_db.json')  # Load your cable TV data

# Remove or modify the `_id` field
if '_id' in cable_tv_df.columns:
    cable_tv_df.drop(columns=['_id'], inplace=True)  # Remove the '_id' column

# Convert to dictionary format
cable_tv_data = cable_tv_df.to_dict(orient='records')

# Insert into MongoDB
cable_tv_collection.insert_many(cable_tv_data)
```

```python
# Fetch and print some documents from the cable_tv_data collection to verify insertion
for record in cable_tv_collection.find().limit(5):  # Limit to 5 records for brevity
    print(record)
```
