```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')  # Adjust URI if using a remote MongoDB instance

# Create or access the database
db = client['media_consumption']

# Create collections
streaming_collection = db['streaming_content']
pricing_collection = db['pricing_data']
```

```python
# Import pandas if not already
import pandas as pd

# Load cleaned data
streaming_data = pd.read_csv('data/cleaned_streaming_content.csv')
pricing_data = pd.read_csv('data/cleaned_all_services_price_history.csv')

# Convert DataFrames to dictionaries and insert into MongoDB
streaming_collection.insert_many(streaming_data.to_dict('records'))
pricing_collection.insert_many(pricing_data.to_dict('records'))
```

```python
# Check document counts
streaming_count = streaming_collection.count_documents({})
pricing_count = pricing_collection.count_documents({})
print(f'Streaming Collection Count: {streaming_count}')
print(f'Pricing Collection Count: {pricing_count}')
```

```python
# Create an index on a common query field, e.g., 'title'
streaming_collection.create_index([('title', 1)])
```
